```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def image_bin(img):
    img1d=img.flatten()
    msg=''
    for pix in img1d:
        msg+="{0:08b}".format(pix)
    return msg
```

```python
def bin_image(msg,r,c):
    pixarr=[]
    i=0
    while i<len(msg):
        pix=int(msg[i:i+8], 2)
        pixarr.append(pix)
        i=i+8
    wm_img=np.array(pixarr).reshape(r,c)
    return wm_img    
    
```

```python
def crop(img,size):    
    xc=img.shape[0]//2
    yc=img.shape[1]//2
    img[xc-size//2:xc+size//2,yc-size//2:yc+size//2]=255
    return img
    
    
```

```python
def norm_data(data):
    """
    normalize data to have mean=0 and standard_deviation=1
    """
    mean_data=np.mean(data)
    std_data=np.std(data, ddof=1)
    #return (data-mean_data)/(std_data*np.sqrt(data.size-1))
    return (data-mean_data)/(std_data)


def ncc(data0, data1):
    """
    normalized cross-correlation coefficient between two data sets

    Parameters
    ----------
    data0, data1 :  numpy arrays of same size
    """
    return (1.0/(data0.size-1)) * np.sum(norm_data(data0)*norm_data(data1))
```

```python
def cal_mse(cover,stego):
    cover=cover.astype('float64')
    stego=stego.astype('float64')
    mse=0
    for i in range(cover.shape[0]):
        for j in range(cover.shape[1]):
                mse=mse+(cover[i,j]-stego[i,j])*(cover[i,j]-stego[i,j])
    mse=mse/(cover.size)
    return mse  
```

```python
def embedding(cover,wmimg):
    r,c=wmimg.shape
    rows,cols=cover.shape
    watermark=image_bin(wmimg)
    cover1d=cover.flatten()
    i=0
    k=0
    while(i<len(watermark)):
        wm=watermark[i:i+2]
        i=i+2
        pix=cover1d[k]
        lpix=list("{0:08b}".format(pix))
        lpix[7]=wm[1]
        lpix[6]=wm[0]
        spix=''.join(lpix)
        cover1d[k]=int(spix,2)
        k=k+1
    stego=np.array(cover1d).reshape(rows,cols)
    return stego 
    
```

```python
def extraction(stego,r,c):
    wmlen=r*c*8
    stego1d=stego.flatten()
    i=0
    k=0
    msg=''
    while(i<wmlen):
        pix=stego1d[k]
        pixbin="{0:08b}".format(pix)
        msg+=pixbin[6:]
        i=i+2
        k=k+1
    rec_wm=bin_image(msg,r,c)
    return rec_wm     
    
```

```python
cover=cv2.imread('./CoverImages/lena.tiff',0)
wm=cv2.imread('./CoverImages/splash.tiff',0)
wm_img=cv2.resize(wm,(64,64))
```

```python
plt.subplot(121)
plt.imshow(cover,cmap='gray')
plt.subplot(122)
plt.imshow(wm_img,cmap='gray')
plt.show()
```

```python
stego=embedding(cover,wm_img)
```

```python
rec_wm=extraction(stego,64,64)
```

```python
plt.subplot(121)
plt.imshow(stego,cmap='gray')
plt.subplot(122)
plt.imshow(rec_wm,cmap='gray')
plt.show()
```

```python
ps=cv2.PSNR(cover,stego)
print(ps)
```

```python
nc=ncc(wm_img,rec_wm)
print(nc)
```

```python
ms=cal_mse(cover,stego)
print(ms)
```
